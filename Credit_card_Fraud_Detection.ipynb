```python
#Importing dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score, accuracy_score
```

```python
df=pd.read_csv(r"creditcard.csv\creditcard.csv")
```

```python
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
fraud.Amount.describe()
```

Since number of legit and fraud transactions are uneven, we make a sample data with same number of legit as fraudulent transactions

```python
legit_evened=legit.sample(n=492)
```

```python
df1=pd.concat([legit_evened,fraud], axis=0)
```

```python
df1.shape
```

```python
df1.Class.value_counts()
```

```python
df1.groupby("Class").mean()
```

```python
x=df1.drop(columns="Class",axis=1)
y=df1["Class"]
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, stratify=y, random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
#for training data
x_pred=lr.predict(x_train)
acc=accuracy_score(x_pred,y_train)
print(acc*100)
```

```python
#for testing data
x_test_pred=lr.predict(x_test)
acc_test=accuracy_score(x_test_pred,y_test)
print(acc_test*100)
```
